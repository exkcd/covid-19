```python
import pandas as pd
import numpy as np
```

```python
covid = pd.read_csv('data/Covid Data.csv')
covid.info()
covid.head()
```

```python
#Removing all rows that contain missing data in binary columns
covid_clean = covid
cols = ['INTUBED', 'PNEUMONIA', 'PREGNANT', 'DIABETES', 'COPD', 'ASTHMA', 'INMSUPR', 'HIPERTENSION', 'OTHER_DISEASE', 'CARDIOVASCULAR', 'OBESITY', 'RENAL_CHRONIC', 'TOBACCO']
for col in cols:
    covid_clean = covid_clean[covid_clean[col] < 3]
covid_clean.info()
```

```python
#Transforming DATE_DIED column into a binary column that just displays whether or not the patient died
covid_clean['PATIENT_DIED'] = np.where(covid_clean['DATE_DIED'] == '9999-99-99', 2, 1)
covid_clean_transform = covid_clean.drop('DATE_DIED', axis = 1)
covid_clean_transform['PATIENT_DIED'].value_counts()
```

```python
covid_clean_transform.info()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = covid_clean_transform.drop(["PATIENT_TYPE", "PATIENT_DIED"], axis=1)
y = covid_clean_transform[["PATIENT_TYPE", "PATIENT_DIED"]]
```

```python
X_train_bin, X_test_bin, y_train_bin, y_test_bin = train_test_split(X, y, test_size = 0.3, random_state=42)
```
